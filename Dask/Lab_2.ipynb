```python
# Listing 4.1
import dask.dataframe as dd
from dask.diagnostics import ProgressBar

fy14 = dd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014.csv')
fy15 = dd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2015.csv')
fy16 = dd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2016.csv')
fy17 = dd.read_csv('Parking_Violations_Issued_-_Fiscal_Year_2017.csv')
fy17
```

```python
# Listing 4.2

fy17 = pd.read_csv("Parking_Violations_Issued_-_Fiscal_Year_2017.csv")
fy17.columns
```

```python
# Listing 4.3
# Import for Python 3.x
from functools import reduce

columns = [set(fy14.columns),
    set(fy15.columns),
    set(fy16.columns),
    set(fy17.columns)]
common_columns = list(reduce(lambda a, i: a.intersection(i), columns))
common_columns
```

```python
# Listing 4.4
fy17[common_columns].head()
```
